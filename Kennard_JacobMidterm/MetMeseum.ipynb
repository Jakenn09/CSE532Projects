### Problem 2
1. Created a copy from the original data frame called df_prob2
2. Used the pandas .dropna method on the new dataframe. Using this method, we pass arguements to determine which values are removed from the dataframe. Problem 2 says we remove all. 
3. The new dataframe is then displayed to the screen, and shows a reduction from 44 columns to 35 columns removing all the columns that have all their data missing. 

```python
df_prob2 = df_prob1.copy() # creates copy of the original dataframe
df_prob2.dropna(axis=1, how='all', inplace=True) # removes all the columns that have all their data missing

df_prob2 # displays a reduction from 44 columns to 35 
```

### Problem 3
1. Created a copy from the previous dataframe which will be used for this problem.
2. Filtered the dataframe using a boolean mask. 'Is Public Domain' is used as the column and each element in that comlumn is compated to True. Only the rows where the condition is True are kept in the  new dataframe.
3. The new dataframe is then displayed to the screen, which shows a reduction in the number of rows which were removed for being False.

```python
df_prob3 = df_prob2.copy # creates copy of the original dataframe
df_prob3 = df_prob2[df_prob2["Is Public Domain"] == True] # filters the dataframe using a boolean mask
df_prob3 # displays a reduction in the number of rows



```

### Problem 4
1. Created a copy from the pervious dataframe which is used in this probelem
2. Used the .fillna() method to replace all the na values in the "Artist Role" column of the new dataframe. This was done due to an error which occured when using the .contains() method that does not accept NA/NaN values.
3. Filtered the rows based on weather the "Artist Role" column contains the string "Maker". Only rows where the condition is true are shown in the new dataframe which is displayed with an updated row count.

```python
df_prob4 = df_prob3.copy() # creates copy of the original dataframe
df_prob4["Artist Role"] = df_prob4["Artist Role"].fillna("") # replaces all the the missing values in "Artist Role" column of the new dataframe with ""

df_prob4 = df_prob4[df_prob4["Artist Role"].str.contains("Maker")] # filters the dataframe based on weather the "Artist Role" column contains the string "Maker"

df_prob4 # displays a reduction in the number of rows 

```

### Problem 5
1. To begin this problem, I imported the numpy library. which is used to import the .where() method which is discussed further down.
2. Created a copy of the previous dataframe which will be used in this problem.
3. Created a new column in our dataframe called "DecadesToMake" and assigned it using basic arithmatic that calculates the difference in years between "Object Begin Date" and "Object End Date" using those column names. 
4. Floor division is used after the difference in years is calcualted to divide the two dates by ten, effectively converting our number to decades, and returns the number as an integer type to avoid floats.
5. Next I wanted to replace any of the values that did not contain an "Object End Date" that would result in the difference being negative. To do this, I utilized the numpy library np.where function to replace all the negative values with a 0. This is done by passing 3 arguements. This function iterates through the "DecadesToMake" and looks for any value less than 0 in df_prob5["DecadesToMake"], replaces it with a 0 in that location, and all values greater than 0 remain the same.

```python
import numpy as np
df_prob5 = df_prob4.copy() # creates copy of the original dataframe

df_prob5["DecadesToMake"] = ((df_prob5["Object End Date"] - df_prob5["Object Begin Date"]) // 10).astype(int) # calculates the difference in years between "Object Begin Date" and "Object End Date" 

df_prob5["DecadesToMake"] = np.where(df_prob5["DecadesToMake"] < 0, 0, df_prob5["DecadesToMake"]) # replaces all the negative values with a 0 using the np.where() function

df_prob5 # displays the new dataframe with added DecadesToMake column


        
```

### Problem 6
1. To begin this problem i imported mathplotlib.pyplot as plt which is used to plot the data for the program. 
2. Next i needed to calculate the number of occurances each classification in the "Classification" comlumn of the dataframe. This was done using the .values_counts() method which returns a Series with the number of unique occurances for each classification.
3. This series was then sorted using the .sort_values() method and makes the acending arguement false to sort the series in decending order so that the most frequent classification is at the top of the Series
4. In order to get the top 10 classifications, I used the .head(10) method which returns the first 10 rows of the series.
5. The plt figsize is adjused for readability and the x and y-axis are labled clearly. 

```python
import matplotlib.pyplot as plt 

class_counts = df_prob5["Classification"].value_counts() # .value_counts() returns a Series containing count of all unique values in df_prob5["Classification"].
sorted_counts = class_counts.sort_values(ascending=False) # .sort_values() sorts the Series in decending order so that the most frequent classification is first.

top_ten_counts = class_counts.head(10) # .head(10) returns the first 10 rows of the Series.

plt.figure(figsize=(7,7))
top_ten_counts.plot(kind='bar', color='red')
plt.title("Top 10 Classifications in df_prob5")
plt.xlabel("Classifications")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
